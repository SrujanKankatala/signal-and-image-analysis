```
#importing different libraries 
import matplotlib.pyplot as plt
import pandas as pd
import math 
import cmath
import numpy as np
df= pd.read_csv('signal_assignment1.csv')
```

Plot x[n] vs n

```
X=[i for i in range(101)] #creating a list of numbers
Y=list(df['x[n]']) #dataframe to list
plt.plot(X,Y)
```

```
X = np.array(X)
def synthesis(Omega, x,y):  #function
  z=0
  for i in x:     #for loop along n
    com=cmath.rect(1,Omega*i)
    z+=y[i]*com
  return cmath.polar(z)[0]
Y_freq= np.arange(-1*math.pi,1*math.pi,0.01*math.pi) #capital_omega
list_1=[]          #creating empty list
for i in Y_freq:
  list_1.append(synthesis(i,X,Y))
```

```
plt.plot(Y_freq,list_1) 
```
